##### 5. Consider the fitted values that result from performing linear regression without an intercept. In this setting, the ith fitted value takes the form $$ \hat y_i = x_i \hat\beta $$ where $$ \hat\beta = (\sum_{i=1}^n x_i y_i) / (\sum_{i'=1}^n {x^2}_{i'} ) $$ Show that we can write $$ \hat y_i = \sum_{i'=1}^n a_{i'} y_{i'} $$ What is $a_{i'}$ ? 

Note: We interpret this result by saying that the fitted values from linear regression are linear combinations of the response values.

|Answer: | |
|--------|----------|

To show that we can write $ \hat{y}_i = \sum_{i'=1}^n a_{i'} y_{i'} $, we need to express $ \hat{y}_i $ in terms of $ y_{i'} $.

Given:
$ \hat{y}_i = x_i \hat{\beta} $
$ \hat{\beta} = \frac{\sum_{i=1}^n x_i y_i}{\sum_{i'=1}^n {x_{i'}}^2} $

Substituting the expression for $ \hat{\beta} $ into $ \hat{y}_i $, we get:
$ \hat{y}_i = x_i \left( \frac{\sum_{i=1}^n x_i y_i}{\sum_{i'=1}^n {x_{i'}}^2} \right) $

This can be rewritten as:
$ \hat{y}_i = \frac{x_i}{\sum_{i'=1}^n {x_{i'}}^2} \sum_{i=1}^n x_i y_i $

Now, let's define $ a_{i'} = \frac{x_{i'}}{\sum_{i'=1}^n {x_{i'}}^2} $. Then, we can express $ \hat{y}_i $ as:
$ \hat{y}_i = \sum_{i'=1}^n a_{i'} y_{i'} $

Therefore, $ a_{i'} = \frac{x_{i'}}{\sum_{i'=1}^n {x_{i'}}^2} $.

##### 7. It is claimed in the text that in the case of simple linear regression of Y onto X, the R2 statistic (3.17) is equal to the square of the correlation between X and Y (3.18). Prove that this is the case. For simplicity, you may assume that $\bar x = \bar y = 0$.





|Answer: | |
|--------| ----------------|


To Prove: $R^2 = Cor(X,Y)^2$
When $\bar x = \bar y = 0 $ 
\begin{align}
Cor(X,Y)^2 = \frac{(\sum_{i=1}^n x_i y_i)^2}{\sum_{i=1}^n x_{i}^2 y_{i}^2} \tag{eq 1}
\end{align}
We know that $TSS = \sum_{i=1}^n (y_i - \bar y_i)^2 = \sum_{i=1}^n (y_i)^2$
and $RSS = \sum_{i=1}^n (y_i - \hat y_i)^2 $

\begin{align}
R^2 &= 1- RSS/TSS \
&= \frac{\sum_iy_i^2 - \sum_i(y_i - x_i\frac{\sum_j{x_jy_j}}{\sum_jx_j^2})^2}
            {\sum_iy_i^2} \
&= \frac{\sum_iy_i^2 - \sum_i(
          y_i^2 -
          2y_ix_i\frac{\sum_j{x_jy_j}}{\sum_jx_j^2} +
          x_i^2 (\frac{\sum_j{x_jy_j}}{\sum_jx_j^2})^2
        )}{\sum_iy_i^2} \

    &= \frac{
          2\sum_i(y_ix_i\frac{\sum_j{x_jy_j}}{\sum_jx_j^2}) -
          \sum_i(x_i^2 (\frac{\sum_j{x_jy_j}}{\sum_jx_j^2})^2)
        }{\sum_iy_i^2} \

    &= \frac{
          2\sum_i(y_ix_i) \frac{\sum_j{x_jy_j}}{\sum_jx_j^2} -
          \sum_i(x_i^2) \frac{(\sum_j{x_jy_j})^2}{(\sum_jx_j^2)^2}
        }{\sum_iy_i^2} \

    &= \frac{
          2\frac{(\sum_i{x_iy_i})^2}{\sum_jx_j^2} -
          \frac{(\sum_i{x_iy_i})^2}{\sum_jx_j^2}
        }{\sum_iy_i^2} \

    &= \frac{(\sum_i{x_iy_i})^2}{\sum_ix_i^2 \sum_iy_i^2} \tag{eq 2}
\end{align}

Hence proved that $R^2 = Cor(X,Y)^2$ from eq 1 and eq 2.

- (d) For the regression of Y onto X without an intercept, the t statistic
for $H_0 : \beta = 0$ takes the form $\hat \beta/SE(\hat \beta)$, where $\hat \beta$ is
given by (3.38), and where $$ SE(\hat \beta) = \sqrt{ \frac {\Sigma_{i=1}^n (y_i - x_i \hat\beta)^2} {(n-1) \Sigma_{i'=1}^n x_{i'}^2} } $$ 

(These formulas are slightly different from those given in Sections
3.1.1 and 3.1.2, since here we are performing regression
without an intercept.) Show algebraically, and confirm numerically
in Python, that the t-statistic can be written as 
$$ \frac{(\sqrt{n-1})\Sigma_{i=1}^n x_i y_i}{\sqrt{(\Sigma_{i=1}^n {x^2}_i)(\Sigma_{i'=1}^n {y^2}_{i'})-(\Sigma_{i'=1}^n {x^2}_{i'} {y^2}_{i'})^2}}$$


Algebraic proof:
Given: $$ \hat{\beta} = \frac{\sum_{i=1}^n x_i y_i}{\sum_{i'=1}^n {x_{i'}}^2}, SE(\hat \beta) = \sqrt{ \frac {\Sigma_{i=1}^n (y_i - x_i \hat\beta)^2} {(n-1) \Sigma_{i'=1}^n x_{i'}^2} } $$

We Know that: 
\begin{align}
t &= \frac{\hat \beta}{SE(\hat \beta)} \
&= \frac{\frac{\sum_{i=1}^n x_i y_i}{\sum_{i'=1}^n {x_{i'}}^2}}{\sqrt{ \frac {\Sigma_{i=1}^n (y_i - x_i \hat\beta)^2} {(n-1) \Sigma_{i'=1}^n x_{i'}^2} }} \
&= \frac{\sum_{i=1}^n x_i y_i}{\sum_{i'=1}^n {x_{i'}}^2} X \frac{\sqrt{{(n-1) \Sigma_{i'=1}^n x_{i'}^2}}}{\sqrt{\Sigma_{i=1}^n (y_i - x_i \hat\beta)^2}}\
&= \frac{\sqrt{n-1} \sum_{i=1}^n x_i y_i}{\sqrt{\sum_{i'=1}^n {x_{i'}}^2 (y_i^2 +(x_i \hat\beta)^2 - 2(y_i)(x_i \hat\beta))}}
\end{align}

Expanding and substituting our expression for $\hat{\beta}$ the denominator becomes: 
$$\big(\sum\limits_{i=1}^n x_{i}^2\big)\Big(\sum\limits_{i'=1}^n y_{i'}^2 - 2 x_{i'}y_{i'}\hat{\beta} + x_{i'}^2 \hat{\beta}^2\Big) = \Big(\sum\limits_{i=1}^n x_{i}^2\Big)\Big(\sum\limits_{i'=1}^n y_{i'}^2\Big) - 2\sum\limits_{i=1}^n x_{i}^2 \sum\limits_{i'=1}^n x_{i'}y_{i'}\frac{\sum\limits_{i=1}^n x_{i}y_{i}}{\sum\limits_{i'=1}^n x_{i'}^2} + \sum\limits_{i=1}^n x_{i}^2 \sum\limits_{i'=1}^n x_{i'}^2 \Bigg(\frac{\sum\limits_{i=1}^n x_{i}y_{i}}{\sum\limits_{i'=1}^n x_{i'}^2}\Bigg)^2$$

Simplifying down to: $$\Big(\sum\limits_{i=1}^n x_{i}^2\Big)\Big(\sum\limits_{i'=1}^n y_{i'}^2\Big) - \Big(\sum\limits_{i'=1}^n x_{i'}y_{i'}\Big)^2$$

Therefore  $$ t = \frac{(\sqrt{n-1})\Sigma_{i=1}^n x_i y_i}{\sqrt{(\Sigma_{i=1}^n {x^2}_i)(\Sigma_{i'=1}^n {y^2}_{i'})-(\Sigma_{i'=1}^n {x^2}_{i'} {y^2}_{i'})^2}}$$
