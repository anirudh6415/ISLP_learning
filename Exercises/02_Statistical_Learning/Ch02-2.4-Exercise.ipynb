Suppose we wish to use this data set to make a prediction for Y when
X1 = X2 = X3 = 0 using K-nearest neighbors.

(a) Compute the Euclidean distance between each observation and
the test point, X1 = X2 = X3 = 0.

(b) What is our prediction with K = 1? Why?

(c) What is our prediction with K = 3? Why?

(d) If the Bayes decision boundary in this problem is highly nonlinear,
then would we expect the best value for K to be large or
small? Why?

### Answer:

````python
import numpy as np
import pandas as pd

d = {'X1': pd.Series([0,2,0,0,-1,1]),
     'X2': pd.Series([3,0,1,1,0,1]),
     'X3': pd.Series([0,0,3,2,1,1]),
     'Y': pd.Series(['Red','Red','Red','Green','Green','Red'])}
df = pd.DataFrame(d)
df.index = np.arange(1, len(df) + 1)
df
````

````python
from math import sqrt
df['distance']=np.sqrt(df['X1']**2+df['X2']**2+df['X3']**2)
df
````

````python
df.sort_values(['distance'])
````

### when k =1, our prediction would be Green, since the nearest neighbour would be in distnace 1.414
### when K = 3 , our prediction would be Red, since the mode is 3 neighbhour , the red has upper hand with 2 reds

In the context of k-nearest neighbors (KNN) classification, the choice of the value for K can have an impact on the decision boundary. The Bayes decision boundary represents the optimal decision boundary based on the true underlying distribution of the data.

If the Bayes decision boundary is highly nonlinear, meaning that the true relationship between the features and the classes is complex and not easily approximated by a simple linear model, then it might be beneficial to use a smaller value for K. Here's why:

Smaller K (e.g., K=1): This allows the algorithm to capture more local variations and adapt to the nonlinearities in the data. Each point's class is determined by its closest neighbor, leading to a more flexible and complex decision boundary.

Larger K (e.g., K=10 or more): This could smooth out the decision boundary by considering a larger number of neighbors, potentially leading to a more linear or smoothed decision surface. This might not capture intricate nonlinearities in the Bayes decision boundary.

In summary, if the Bayes decision boundary is highly nonlinear, a smaller value for K in KNN might be more appropriate as it allows the model to better capture the local complexities in the data. However, the optimal value for K depends on the specific characteristics of the data, and it's often determined through cross-validation or other model selection techniques.
